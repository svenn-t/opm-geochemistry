# Equilibrium between a solution and gas phase
This example demonstrates equilibration between an aqueous solution, a mineral assemblage, and a constant-volume gas phase. $\mathrm{CO_2(g)}$ is generated by mineral reactions, and gas pressure evolves freely as an equilibrium outcome.

<!-- Equation labels as ordinary links -->
<div id="_auto1"></div>

$$
\begin{equation}
\text{CaMg(CO$_{3}$)$_2$(s)} + 2\text{SiO}_2 {\rightleftharpoons} \text{CaMg(Si$_{2}$)O$_6$(s)} + 2\mathrm{CO_2(g)}.
\label{_auto1} \tag{1}
\end{equation}
$$

**Input pressure and partial pressure.**

In this example the pressure is unknown before the calculation starts. `Pres` keyword is only used as an *initial guess*. Furthermore the partial pressures given in the `GAS_PHASE` as inputs are converted to moles using the ideal gas law. However, the final gas partial pressures are calculated using the Peng Robinson equation of state for mixtures [[appelo2014equations;@ peng1976new]](#appelo2014equations;@ peng1976new).

## Input file

        GEOCHEM
        Temp	350
        Pres	300e5
        
        SOLUTION
        pH	5 charge
        /end
        
        EQUILIBRIUM_PHASES
        dolomite 1 0
        quartz 1 0
        diopside 1 0
        /end
        
        GAS_PHASE
        volume_frac	1.33
        CO2,g	0
        H2O,g	0
        Ndg,g	10e5
        /end
        /end


**Explanation of input file:**

`GEOCHEM .. /end`:
  :    
  Defines a GEOCHEM calculation block.

`Temp`:
  :    
  temperature of solution, units: Celsius, default $\text{350}^\circ\text{C}$

`Pres`:
  :    
  pressure of solution, units: Pascal. NOTE: `Pres` only serves as an *initial guess* for the numerical solver. In a constant-volume gas phase, pressure is not fixed and is allowed to change during equilibration.

`EQUILIBRIUM_PHASES .. /end`:
  :    
  minerals present, here dolomite, quartz and diopside, see more description in the [solution example](../../solution/html/solution-readable.html).  

`GAS_PHASE ../end`:
  :    

  * `volume_frac`: ratio of gas volume to water volume

  * $\mathrm{CO_2(g)}$ and $\mathrm{H_2O(g)}$ start at zero, any $\mathrm{CO_2(g)}$ or $\mathrm{H_2O(g)}$ is generated by equilibration.

  * `Ndg,g` redox decoupled nitrogen gas (see below) at initial partial pressure of 10 bar. 

**Redox decoupled gas.**

In this example, the gas phase includes an inert species (`Ndg`) that does not participate in aqueous speciation or redox reactions. Conceptually, `Ndg` represents a pressure-carrying gas such as $\mathrm{N_2(g)}$, but without explicit redox coupling to the aqueous system. This is because the basis specie is $\text{NO}_3^{-}$ and by default the equilibrium reaction for $\mathrm{N_2(g)}$ is
$\mathrm{N_2(g)}{\rightleftharpoons} 12\text{H}^+ -6 \text{H}_2\text{O} + 2 \text{NO}_3^- + 10 e^-$. This is not a reasonable modeling assumption as nitrogen is for all practical purposes an inert gas.

## Description of output
The pH of the solution is close to 5, due to
* carbonate reactions associated with dolomite

* Dissolution–precipitation equilibria involving $\mathrm{CO_2}$ and bicarbonate

Major aqueous species
* $\mathrm{Ca^{2+}}$ and $\mathrm{Mg^{2+}}$ Released from partial dolomite breakdown during equilibration.

* $\mathrm{HCO_3^-}$ The primary dissolved carbon species, reflecting equilibration between aqueous carbonate chemistry and the $\mathrm{CO_2}$-rich gas phase.

* $\text{SiO}_2$ Controlled by equilibrium with quartz and diopside.

The final pressure close to 240 bar, and the gas phase composition is in the file `<root_name>_gas_phase.out`

**Key observations:**
1. Water vapor dominates the gas phase. At $\text{350}^\circ\text{C}$, water readily partitions into the gas phase. Because the gas volume is fixed, evaporation of water leads to a large contribution to total pressure. This behavior is expected and physically realistic for high-temperature fluid–gas systems.

2. $\mathrm{CO_2}$ is produced by mineral equilibration. The presence of ~27 mol  $\mathrm{CO_₂}$ reflects production via the buffered reaction in the equation at the top. $\mathrm{CO_2}$ accumulates in the gas phase until its fugacity satisfies the mineral equilibrium, at which point further decarbonation is inhibited.

**Notice.**

$\mathrm{CO_2}$  pressure is an output, not an imposed condition. Its final value reflects coupling between mineral stability and gas volume

# Appendix: Gas buffers and Peng-Robinson parameters
For a (real) gas the thermodynamic activity is measured relative to an reference state of atmospheric pressure $p_0=1$ atm, e.g. for hydrogen gas

<!-- Equation labels as ordinary links -->
<div id="_auto2"></div>

$$
\begin{equation}
a_\mathrm{H_2(g)}=\frac{f_\mathrm{H_2(g)}}{p_0},
\label{_auto2} \tag{2}
\end{equation}
$$

where $f$ is called the *fugacity*. For an ideal gas the fugacity is simply the partial pressure (i.e. the pressure the gas would exert if it was the only gas in a container), for real gasses at a (mechanical) pressure of $p$, the fugacity is different $f=\phi p$. $\phi$ is called a *fugacity coefficient*. The fugacity coefficient are in many cases estimated from [[appelo2014equations]](#appelo2014equations) (Note that in the paper the last term in the equation below has a `+` term, this is not correct)

<!-- Equation labels as ordinary links -->
<div id="_auto3"></div>

$$
\begin{equation}
\ln \phi=\left(\frac{P V_m}{R T}-1\right)-\ln\left(\frac{P(V_m-b)}{RT}\right)
-\frac{a\alpha(T)}{2.828 b R T}\ln\left(\frac{V_m+2.414 b}{V_m-0.414b}\right),
\label{_auto3} \tag{3}
\end{equation}
$$

where the molar volume, $V_m$, and pressure $P$ is determined from the Peng-Robinson equation of state (EOS)

<!-- Equation labels as ordinary links -->
<div id="_auto4"></div>

$$
\begin{equation}
P = \frac{RT}{V_m-b}-\frac{\alpha(T) a}{V_m^2+2bV_m-b^2}.
\label{_auto4} \tag{4}
\end{equation}
$$

The parameters are

<!-- Equation labels as ordinary links -->
<div id="_auto5"></div>

$$
\begin{equation}
a=0.457235 \frac{(RT_c)^2}{p_c}
\label{_auto5} \tag{5}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto6"></div>

$$
\begin{equation}  
b=0.077796\frac{RT_c}{p_c}
\label{_auto6} \tag{6}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto7"></div>

$$
\begin{equation}  
\alpha(T)=\left(1+(0.37464+1.54226\omega -0.26992\omega^2)\left(1-\sqrt{\frac{T}{T_c}}\right)\right)^2	
\label{_auto7} \tag{7}
\end{equation}
$$

$p_c$ and $T_c$ are the critical pressure and temperature respectively. Below are the values listed that are already included in `GeoChemX`. Note that the three last entries are *redox decoupled* versions of hydrogen, hydrogen sulfide and nitrogen gas. Wh
